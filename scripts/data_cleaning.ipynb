```python
path = r"../data/coronavirusdataset.csv"
data = pd.read_csv(path)
df = pd.DataFrame(data)
# print(data)
```

```python
print(data.info())
```

```python
print(df.columns)
```

```python
# droping most empty values 
df = df.drop(columns=['batch_date','rapid_strep_results','rapid_flu_results','cough_severity','sob_severity','cxr_findings', 'cxr_impression', 'cxr_label',
       'cxr_link'])
print(df)
```

```python
# print(df.columns)
df = df.dropna()
df.info()
```

```python
y = df["covid19_test_results"]
X = df.drop(columns=["covid19_test_results"])
```

```python
# check labels
print(y)
```

```python
# check features
print(X)
```

```python
print(X.info())
```

```python
cols = X.columns
X_encoded = X.copy()
le = LabelEncoder()
for col_name in X_encoded.select_dtypes(include=["object","string"]):
    
    X_encoded[col_name]=le.fit_transform(X_encoded[col_name])

y_encode =le.fit_transform(y)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_encoded,y_encode)
model = SVC(kernel='rbf').fit(X_train,y_train)
```

```python
print(f" training score {model.score(X_train,y_train)}")
print(f" test score {model.score(X_test,y_test)}")
```
